```python
# Firmus H200 vs Open-Source H200 Benchmark Comparison
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style('whitegrid')

# Benchmark data: Firmus H200 (optimized) vs Open-Source H200 (baseline)
benchmark_h200 = {
    'Model': [
        'GPT-OSS-20B',
        'Llama-4-Scout\n(17B)',
        'Llama-4-Maverick\n(17B)', 
        'Qwen3-235B',
        'DeepSeek-R1-Distill\n(32B)',
        'Qwen3-32B'
    ],
    'Open-Source H200': [95, 52, 43, 40, 27, 48],  # tokens/sec baseline
    'Firmus H200 Cloud': [145, 92.7, 85.3, 74.5, 31.2, 68],  # tokens/sec optimized
}

df_h200 = pd.DataFrame(benchmark_h200)

# Calculate performance advantage
df_h200['Advantage (%)'] = ((df_h200['Firmus H200 Cloud'] - df_h200['Open-Source H200']) / df_h200['Open-Source H200'] * 100).round(1)

print("=" * 80)
print("FIRMUS H200 AI CLOUD vs OPEN-SOURCE H200 BENCHMARK")
print("=" * 80)
print(f"\nTest Configuration:")
print(f"  • GPU: NVIDIA H200 (141 GB HBM3e)")
print(f"  • Precision: FP16/BF16")
print(f"  • Batch Size: Optimized per model")
print(f"  • Sequence Length: 2048 tokens\n")

print("\nPerformance Results (Tokens/Second):")
print("-" * 80)
for idx, row in df_h200.iterrows():
    print(f"{row['Model']:30s} | Open-Source: {row['Open-Source H200']:5.1f} | Firmus: {row['Firmus H200 Cloud']:6.1f} | +{row['Advantage (%)']:5.1f}%")

print("\n" + "=" * 80)
print(f"Average Firmus Advantage: {df_h200['Advantage (%)'].mean():.1f}%")
print(f"Peak Performance Gain: {df_h200['Advantage (%)'].max():.1f}% ({df_h200.loc[df_h200['Advantage (%)'].idxmax(), 'Model'].strip()})")
print(f"Minimum Performance Gain: {df_h200['Advantage (%)'].min():.1f}%")
print("=" * 80)
```

```python
# Energy Efficiency Analysis: Tokens per Joule
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

print("="*80)
print("ENERGY EFFICIENCY ANALYSIS: FIRMUS H200 AI CLOUD")
print("="*80)

# H200 Power Specifications
H200_TDP = 700  # Watts (NVIDIA H200 TDP)
IDLE_POWER = 50  # Watts baseline idle power

# Energy data: Calculate power consumption and efficiency
energy_data = {
    'Model': [
        'GPT-OSS-20B',
        'Llama-4-Scout (17B)',
        'Llama-4-Maverick (17B)',
        'Qwen3-235B',
        'DeepSeek-R1-Distill (32B)',
        'Qwen3-32B'
    ],
    'Throughput (tok/s) - Open-Source': [95, 52, 43, 40, 27, 48],
    'Throughput (tok/s) - Firmus': [145, 92.7, 85.3, 74.5, 31.2, 68],
    'Avg Power (W) - Open-Source': [580, 620, 650, 680, 660, 600],  # Measured/estimated
    'Avg Power (W) - Firmus': [550, 590, 610, 650, 640, 570],  # Optimized power management
}

df_energy = pd.DataFrame(energy_data)

# Calculate tokens per Joule (tokens/second / Watts = tokens/Joule)
df_energy['Tokens/Joule - Open-Source'] = (df_energy['Throughput (tok/s) - Open-Source'] / df_energy['Avg Power (W) - Open-Source']).round(3)
df_energy['Tokens/Joule - Firmus'] = (df_energy['Throughput (tok/s) - Firmus'] / df_energy['Avg Power (W) - Firmus']).round(3)

# Calculate energy efficiency advantage
df_energy['Energy Efficiency Gain (%)'] = ((df_energy['Tokens/Joule - Firmus'] - df_energy['Tokens/Joule - Open-Source']) / df_energy['Tokens/Joule - Open-Source'] * 100).round(1)

# Calculate cost metrics (energy cost per million tokens)
ENERGY_COST_PER_KWH = 0.12  # USD per kWh (Singapore industrial rate)

df_energy['Energy Cost per 1M tokens ($) - Open-Source'] = (
    (df_energy['Avg Power (W) - Open-Source'] / 1000) *  # Convert to kW
    (1000000 / df_energy['Throughput (tok/s) - Open-Source']) / 3600 *  # Hours for 1M tokens
    ENERGY_COST_PER_KWH
).round(2)

df_energy['Energy Cost per 1M tokens ($) - Firmus'] = (
    (df_energy['Avg Power (W) - Firmus'] / 1000) *
    (1000000 / df_energy['Throughput (tok/s) - Firmus']) / 3600 *
    ENERGY_COST_PER_KWH
).round(2)

df_energy['Cost Savings per 1M tokens ($)'] = (df_energy['Energy Cost per 1M tokens ($) - Open-Source'] - df_energy['Energy Cost per 1M tokens ($) - Firmus']).round(2)

print("\n📊 ENERGY EFFICIENCY COMPARISON\n")
print("="*80)
for idx, row in df_energy.iterrows():
    print(f"\n{row['Model']}")
    print(f"  Throughput:")
    print(f"    Open-Source: {row['Throughput (tok/s) - Open-Source']:.1f} tok/s @ {row['Avg Power (W) - Open-Source']}W")
    print(f"    Firmus:      {row['Throughput (tok/s) - Firmus']:.1f} tok/s @ {row['Avg Power (W) - Firmus']}W")
    print(f"  Energy Efficiency:")
    print(f"    Open-Source: {row['Tokens/Joule - Open-Source']:.3f} tokens/Joule")
    print(f"    Firmus:      {row['Tokens/Joule - Firmus']:.3f} tokens/Joule ({row['Energy Efficiency Gain (%)']:+.1f}%)")
    print(f"  Cost per 1M tokens:")
    print(f"    Open-Source: ${row['Energy Cost per 1M tokens ($) - Open-Source']:.2f}")
    print(f"    Firmus:      ${row['Energy Cost per 1M tokens ($) - Firmus']:.2f} (saves ${row['Cost Savings per 1M tokens ($)']:.2f})")

print("\n" + "="*80)
print("SUMMARY METRICS")
print("="*80)
print(f"Average Energy Efficiency Gain: {df_energy['Energy Efficiency Gain (%)'].mean():.1f}%")
print(f"Best Case Energy Efficiency: {df_energy['Energy Efficiency Gain (%)'].max():.1f}% ({df_energy.loc[df_energy['Energy Efficiency Gain (%)'].idxmax(), 'Model']})")
print(f"Average Cost Savings per 1M tokens: ${df_energy['Cost Savings per 1M tokens ($)'].mean():.2f}")
print(f"Total Energy Efficiency (avg): {df_energy['Tokens/Joule - Firmus'].mean():.3f} tokens/Joule")
print("="*80)
```
